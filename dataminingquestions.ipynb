```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
import csv
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
# Navigate to the website
driver.get('https://ngware.fr/mf2/revisef2.html')
# Wait for the page to load
time.sleep(2)
```

```python
# Find the search box and enter the job title
epreuve = driver.find_elements(By.ID, 'epreuve')
for item in epreuve:
    item.click()
    time.sleep(0.1)
```

```python
theme = driver.find_elements(By.ID,'theme')
for item in theme:
    item.click()
    time.sleep(0.1)
```

```python
responses = driver.find_elements(By.LINK_TEXT ,"Réponse")
for item in responses:
    item.click()
    time.sleep(0.1)
```

```python
resp = driver.find_elements(By.LINK_TEXT ,"Réponse")
resp[0].find_element(By.XPATH,'./..').find_elements(By.XPATH,'./*')[1].text
```

```python
sections=[]
questions=[]
responses=[]
even=True

for i in driver.find_elements(By.CLASS_NAME,"gris"):
    if even:
        even=False
        pass
    else:
        even=True
        sections.append(i.text)

for i in driver.find_elements(By.CLASS_NAME,"danlebleu"):
    questions.append(i.text)

resp = driver.find_elements(By.LINK_TEXT ,"Réponse")
for i in resp:
    responses.append(i.find_element(By.XPATH,'./..').find_elements(By.XPATH,'./*')[1].text)
```

```python
print(len(sections))
print(len(questions))
print(len(responses))
```

```python
data = {
    'Sec':sections,
    'Questions':questions,
    'Answers':responses
}
df = pd.DataFrame(data)
df
```

```python
new_columns = df['Sec'].str.split(' / ', expand=True)
new_columns.columns = ['Field','Theme','Date','City','Points','File','number','dunno']
new_columns

df1 = pd.concat([new_columns,df.drop(columns=['Sec'])],axis=1)
df2 = df1.drop(columns=['File','number','dunno'])
df2.Questions.replace('\n',' ',regex=True,inplace=True)
df2.Answers.replace('\n',' ',regex=True,inplace=True)
df2.Points.replace(' points','',regex=True,inplace=True)
df2.Answers = df2.Answers.str.strip()
df2
```

```python
import re
def remove_illegal_chars(text):
    # Replace all characters with ASCII values 0-31 and 127 (DEL) using a regex pattern
    if isinstance(text, str):
        return re.sub(r'[\x00-\x1F\x7F]', '', text)
    return text

df2 = df2.applymap(remove_illegal_chars)
```

```python
df2.to_excel('output.xlsx',index=False,header=True)
```

```python
df2.to_csv('output.csv',index=False,header=True,sep=';')
```
